The first step is to import the useful packages: numpy, matplotlib.pyplot and the third one is mpl_toolkits.mplot3d for plotting 3D figures. The one thing to note here is we are not importing whole package of mpl_toolkits.mplot3d, we are just importing Axes3D library which is a part of mpl_toolkits.mplot3d.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D     
```

The next step is to load the data set. We open the file data_2d.csv and import the data line by line. We then split each line by the ',' comma. We then assign the sample point before first comma to x1, after the first comma to x2 and the sample point after the second comma to y. Ater this, we convert x1, x2 and y to float and then append x1 and x2 into X and y into Y.

```python
#Loading the data
X = []
Y = []
for line in open('data_2d.csv'):
    x1, x2, y = line.split(',')
    X.append([float(x1), float(x2), 1])
    Y.append(float(y))
```

We then convert the input matrix X and output matrix Y into numpy arrays and reassign them into X and Y respectively.

```python
#Converting X and Y into numpy arrays
X = np.array(X)
Y = np.array(Y)
```

After this, we obtain the 3D figure of X and Y with one dimension being first column of X, second dimension being second column of X and third dimension being Y. As you can see, the scatter plot turns out to be linear.

```python
#Plotting X and Y
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], Y)
plt.show()
```

We now calculate the weights of the multivariate linear regression model using w = inv[(transpose(X))X]transpose(X)Y and then use that weight vector to find the predicted Yhat matrix, Yhat = Xw with X of dimension NxD and w of dimension Dx1.

```python
#Calculating weights of the model
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat = np.dot(X, w)
```

The next step is to calculate the R square of the model. 

```python
#calculating R square of the model
diff1 = Y - Yhat
diff2 = Y - Y.mean()
Rsquare = 1 - (diff1.dot(diff1)/diff2.dot(diff2))
print("The R square of the model is ", Rsquare)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], Y)
ax.plot(sorted(X[:, 0]), sorted(X[:, 1]), sorted(Yhat))
plt.show()
```
