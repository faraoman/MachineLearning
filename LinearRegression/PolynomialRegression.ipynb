```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the data
X = []
Y = []
for line in open('data_poly.csv'):
    x, y = line.split(',')
    x = float(x)
    X.append([1, x, x*x])
    Y.append(float(y))
```

```python
#Convert to numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
#Scatter plot of X and Y
plt.scatter(X[:, 1], Y)
plt.title("Scatter plot of x(second column of X) and Y")
plt.xlabel('X[:, 1]')
plt.ylabel('Y')
plt.show()
```

```python
#Calculating weights
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat = np.dot(X, w)
```

```python
#plotting all of them
plt.scatter(X[:, 1], Y)
plt.plot(sorted(X[:,1]), sorted(Yhat),color='red')
plt.title('Red line: predicted values')
plt.xlabel('X[:, 1]')
plt.ylabel('Y')
plt.show()
```

```python
#calculating R square of the model
diff1 = Y - Yhat
diff2 = Y - Y.mean()
Rsquare = 1 - (diff1.dot(diff1)/diff2.dot(diff2))
print("The R square of the model is ", Rsquare)
```
