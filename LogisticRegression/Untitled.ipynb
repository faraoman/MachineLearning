The first step is to import the required packages.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Set number of onservations, N = 100

```python
No_of_observations = 100   
No_of_Dimensions = 110
```

Generate the input and output matrix as shown below:

```python
X_input = (np.random.random((No_of_observations, No_of_Dimensions))-0.1)*10       
w_dash =  np.array([1, 0.1, -0.1] + [0]*(No_of_Dimensions-3))
```

Next step is calculate the Y corresponding to w_dash

```python
Y_output = X_input.dot(w_dash) + np.random.randn(No_of_observations)*0.1
```

Now, let's apply gradient descent to find the effect of L1 regularization on costs.

```python
costs = []
w = np.random.randn(No_of_Dimensions)/np.sqrt(No_of_Dimensions)
L1_coeff = 10    
learning_rate = 0.001
for t in range(500):
    Yhat = X_input.dot(w)
    delta = Yhat - Y_output
    w = w - learning_rate*(X_input.T.dot(delta) + L1_coeff*np.sign(w))
    meanSquareError = delta.dot(delta)/No_of_observations
    costs.append(meanSquareError)
```

```python
plt.plot(costs)
plt.title("Plot of costs of L1 Regularization")
plt.ylabel("Costs")
plt.show()
```
